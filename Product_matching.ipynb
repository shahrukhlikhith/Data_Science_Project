# Product matching

```python
import numpy as np
import pandas as pd
```

## Import Datasets

```python
data1 = pd.read_csv('flipkart_com-ecommerce_sample.csv')
data2 = pd.read_csv('amz_com-ecommerce_sample.csv')
```

```python
data1.head()
```

```python
data1.isnull().sum()
```

There are missing values in the flipkart dataset. So, we are going to show as 'Currently Unavailable' for the missing values

```python
data2.head()
```

```python
data2.isnull().sum()
```

Here we taking the columns which are necessary.

There are multiple products with same name but the <code>uniq_id</code> is different for each product. So we will use <code>uniq_id</code> column to merge the two tables without getting extra rows.

```python
df1 = data1[['uniq_id','product_name','retail_price','discounted_price']]
df2 = data2[['uniq_id','product_name','retail_price','discounted_price']]
```

```python
df = pd.merge(df1, df2, on = 'uniq_id', how = 'inner')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Changing the name of the columns for the expected output.
df.rename(columns = {"product_name_x": "Product name in Flipkart",
                     "product_name_y": "Product name in Amazon",
                     'retail_price_x':'Retail Price in Flipkart',
                     'retail_price_y':'Retail Price in Amazon',
                     'discounted_price_x':'Discounted Price in Flipkart',
                     'discounted_price_y':'Discounted Price in Amazon'}, inplace = True)
```

```python
#removing 'uniq_id' column
df.drop(columns = ['uniq_id'],inplace = True)
```

```python
df.head()
```

Replacing missing values with '0' until the column changes into <code>int</code> datatype.

```python
df.fillna(0, inplace = True)
```

```python
df[['Retail Price in Flipkart','Discounted Price in Flipkart']] = df[['Retail Price in Flipkart','Discounted Price in Flipkart']].astype('int64')
```

```python
# Replacing 0's back to Unavailable
df.replace(0,'Currently Unavailable', inplace = True)
```

Creating a uppercase product name table, because user may give the input either in lower case or upper case. So we are going to change the user input to upper and use the table to find the user product.

```python
temp1 = df.loc[:, 'Product name in Flipkart'].str.upper()
temp2 = df.loc[:, 'Product name in Amazon'].str.upper()
temp3 = pd.concat([temp1, temp2], axis=1)
temp3.head()
```

```python
df.isnull().sum()
```

## Final Result

```python
from IPython.display import HTML
```

```python
User_product = str(input())
User_product = User_product.upper()
product_fpkt = {}
product_amz = {}
for i in range(df.shape[0]):
    product_fpkt[i] = set(temp3['Product name in Flipkart'][i].split())
    product_amz[i] = set(temp3['Product name in Amazon'][i].split())
product_append = set(User_product.split())
fpkt_product_append = [product_append.issubset(i) for i in product_fpkt.values()]
amz_product_append = [product_append.issubset(i) for i in product_amz.values()]
product_index = [i for i in range(len(amz_product_append)) if (fpkt_product_append[i] or amz_product_append[i])]
if len(product_index):
    final_df = df.iloc[product_index]
    display(HTML(final_df.to_html(index=False)))
else:
    print('Your product is currently unavailable in both flipkart and amazon')
```
